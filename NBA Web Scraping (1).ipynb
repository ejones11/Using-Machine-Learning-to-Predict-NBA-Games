```python
season_list = [
    '1996-97',
    '1997-98',
    '1998-99',
    '1999-00',
    '2000-01',
    '2001-02',
    '2002-03',
    '2003-04',
    '2004-05',
    '2005-06',
    '2006-07',
    '2007-08',
    '2008-09',
    '2009-10',
    '2010-11',
    '2011-12',
    '2012-13',
    '2013-14',
    '2014-15',
    '2015-16',
    '2016-17',
    '2017-18',
    '2018-19',
    '2019-20',
    '2020-21',
    '2021-22'
]
season_types = ['Regular+Season', 'Playoffs']
```

```python
dfs=[]
for st in season_types:
    for season_id in season_list:
        url=f'https://stats.nba.com/stats/teamgamelogs?DateFrom=&DateTo=&GameSegment=&LastNGames=0&LeagueID=00&Location=&MeasureType=Four%20Factors&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=Totals&Period=0&PlusMinus=N&Rank=N&Season={season_id}&SeasonSegment=&SeasonType={st}&ShotClockRange=&VsConference=&VsDivision='
        response = requests.get(url=url, headers=headers).json()
        team_box_scores = response['resultSets'][0]['rowSet']
        nba_df=pd.DataFrame(team_box_scores,columns=column_list)
        print(st, season_id)
        nba_df.insert(1, "Season_Type", st)
        dfs.append(nba_df)
out = pd.concat(dfs, ignore_index=True)
```
